### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
func = 10 * x - 5
d_func = sp.diff(func, x)

print('Похідна від функції 10х-5:', d_func)

x_values = np.linspace(0, 10, 11)
f_func = sp.lambdify(x, func, 'numpy')
df_func = sp.lambdify(x, d_func, 'numpy')

f_list = f_func(x_values)
df_list = df_func(x_values)
if np.isscalar(df_list):
    df_list = np.full_like(x_values, df_list)
#print(f_list, df_list)

plt.plot(x_values, f_list, label='f(x) = 10x-5', color='blue')
plt.plot(df_list, label="f'(x)= 10", color='red',  linestyle='--')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
x_sin = np.linspace(-5,5,100)
f_sin = np.sin(x_sin)

df_sin = sp.diff(sp.sin(x), x)
print(df_sin)

```

```python
d_sin = np.cos(x_sin)

plt.plot(x_sin, f_sin, label='f(x)=sin(x)', color='red')
plt.plot(x_sin, d_sin, label='df(x)=cos(x)', color='green')

plt.title("Графік f(x)=sin(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
```

```python
x = sp.symbols('x')

f_ln = sp.ln(x)
d_ln = sp.diff(f_ln, x)

ln_func = sp.lambdify(x, f_ln, 'numpy')
ln_der = sp.lambdify(x, d_ln, 'numpy')

x_ln = np.linspace(0.1,50,100)

y_ln_func = ln_func(x_ln)
y_ln_der = ln_der(x_ln)

plt.plot(x_ln, y_ln_func, label='ln(x)', color='red')
plt.plot(x_ln, y_ln_der, label='derivative ln(x)', color='blue')
plt.legend()
plt.grid(True)
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

costs = x**2 - 5*x + 1
d_costs = sp.diff(costs, x)

x_costs = np.linspace(-5,20,200)

costs_func = sp.lambdify(x, costs, 'numpy')
costs_der = sp.lambdify(x, d_costs, 'numpy')

f_costs = costs_func(x_costs)
df_costs = costs_der(x_costs)

plt.plot(x_costs, f_costs, label='f(x)=x^2 - 5x + 1', color='red')
plt.plot(x_costs, df_costs, label='df(x)=x^2 - 5x + 1', color='green')

plt.title("Графік f(x)=sin(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
```
